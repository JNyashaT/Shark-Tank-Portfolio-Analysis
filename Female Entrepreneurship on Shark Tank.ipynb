# Value by Industry

```python
tank
```

```python
tank.drop(["OriginalAirDate", "USViewership", "BarbaraCorcoranPresent", "MarkCubanPresent", "LoriGreinerPresent", "RobertHerjavecPresent", "DaymondJohnPresent", "KevinOLearyPresent", "GuestPresent"], axis = 1, inplace = True)
```

```python
tank
```

```python
winners = tank[tank.TotalDealAmount.notna()]
```

```python
gender = ["Male", "Female", "Mixed Team"]
winners["Gender"] = pd.Categorical(winners["PitchersGender"], categories = gender, ordered = True)
categories = winners.groupby(["Industry","PitchersGender"]).size().unstack(fill_value = 0)
```

```python
categories
```

```python
fig, ax = plt.subplots(figsize =(10,5))
scaled = winners["TotalDealAmount"]/1000000
sns.barplot(x = "Industry", y = scaled, data = winners, hue = "PitchersGender", ax=ax, palette = "Set1")
plt.xticks(rotation = 90)
plt.xlabel("Industry")
plt.ylabel("Average Deal Valuation (in millions)")
plt.title("Gender Comparison of Common Industries Highlighted on Shark Tank")
plt.legend(title = "Gender")
plt.savefig('finalfig.jpg', bbox_inches='tight', dpi=100)
```

```python
winners["TotalDealAmount"].describe()
```

```python
worth["hue"] = "female"
worth2["hue"] = "male"
worth3["hue"] = "mixed group"

update = pd.concat([worth, worth2, worth3])
scale_val = update["Portfolio Value"]/1000000
```

```python
update
```

```python
sns.barplot(x = "Shark Name", y = scale_val, hue = "hue", data = update, palette = "Set1")
plt.xticks(rotation = 45)
plt.xlabel("Shark Name")
plt.ylabel("Average Portfolio Value (in millions)")
plt.title("Comparison of Gender-Specific Operated Companies and Average Portfolio Value Per Shark", pad = 20)
plt.legend(title = "Gender")
plt.savefig('sharkfig.jpg', bbox_inches='tight', dpi=150)
```
